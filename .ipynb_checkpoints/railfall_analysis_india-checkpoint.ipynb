# How much will it rain this year?

### India Rainfall Analysis

Mansoon prediction clearly is great importance for india. In this project we will predict the long term rainfall in different division or state of India in advance.We predict the amount of rainfall using past data.Indian meteorological department provides forecasting data required for project.

### Dataset

[Dataset-1](https://data.gov.in/resources/subdivision-wise-rainfall-and-its-departure-1901-2015) have been taken from [data.gov.in](https://data.gov.in/catalog/rainfall-india) which have average rainfall from 1951-2000 for each district, for every month and [Dataset-2](https://data.gov.in/resources/district-rainfall-normal-mm-monthly-seasonal-and-annual-data-period-1951-2000) average rainfall for every year from 1901-2015 for each state.

```python
# import required libraries
import numpy as np   # for algebra calculation
import pandas as pd  # for data pre-processing (reading .CSV file, manipulation of data frame etc.)
import matplotlib.pyplot as plt # for interactive graphs and plots
import seaborn as sns 
```

### Import .CSV file using pandas pd.read_csv function

```python
data = pd.read_csv('data/sub-division_rainfall_act_dep_1901-2015.csv',sep=",")
data = data.fillna(data.mean())
data.info()
```

### Understand the Dataset

 * Above Dataset have 36 subdivision or states and 20 attribute (individual months rain fall data, annual rainfall data and   at last combination of 3 months rain fall data
 * Data is from 1901- 2015 and rainfall measured in MM 

   ##### **Months combination** <br>
   **JF** - jan to Feb <br>
   **MAM** - Mar to May <br> 
   **JJAS** - June to Sep <br>
   **OND** - Oct to Dec

### Select features for taining 

```python
from sklearn.preprocessing import StandardScaler
from sklearn import preprocessing 
from sklearn.pipeline import Pipeline
df2.head(0)
features = []
for element in df2.head(0):
    features.append(element)
features.remove('ANNUAL')
features.remove('JAN')
features.remove('FEB')
features.remove('MAR')
features.remove('APR')
features.remove('OCT')
features.remove('NOV')
features.remove('DEC')
features.remove('AUG')
features.remove('SEP')
print(features)
```

### Target feature (to predict August rainfall)

```python
features2 = ['AUG']
```

```python
# Separating the features
x  = df2.loc[:, features].values
# Separating the target
y = df2.loc[:,features2].values
# Standardizing the features
```

```python
#splitting training and testing data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

# SVM

```python
from sklearn import svm
from sklearn.datasets import make_classification
clf = svm.SVC()
X_train, y_train = make_classification()
X_test, y_test = make_classification()
clf.fit(X_train, y_train)
```

```python
print(clf.intercept_)
```

```python
clf.predict(X_test)
```

```python
clf.score(X_test, y_test, sample_weight=None)
```

```python
y_score = clf.decision_function(X_test)
```

# Random forest

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(max_depth=2, random_state=0)
clf.fit(X_train, y_train)
clf.predict(X_test)
clf.score(X_test, y_test)
```
